```python
import pandas as pd
import os
import matplotlib.pyplot as plt
```

```python
data = os.sep.join(['..', 'ML-1M-tensorboard-data'])
loss_folder = os.sep.join([data, 'loss'])
ndcg_folder = os.sep.join([data, 'ndcg'])
recall_folder = os.sep.join([data, 'recall'])
```

```python
loss_normal = os.sep.join([loss_folder, '03-07-07h27m46s-ml-1m-lgn.csv'])
loss_sorted = os.sep.join([loss_folder, '03-08-06h57m27s-ml-1m_sorted_rating_std-curriculum-3-lgn.csv'])
```

```python
ndcg_normal = os.sep.join([ndcg_folder, '03-07-07h27m46s-ml-1m-lgn_Test_NDCG@[20]_20.csv'])
ndcg_sorted = os.sep.join([ndcg_folder, '03-08-06h57m27s-ml-1m_sorted_rating_std-curriculum-3-lgn_Test_NDCG@[20]_20.csv'])

recall_normal = os.sep.join([recall_folder, '03-07-07h27m46s-ml-1m-lgn_Test_Recall@[20]_20.csv'])
recall_sorted = os.sep.join([recall_folder, '03-08-06h57m27s-ml-1m_sorted_rating_std-curriculum-3-lgn_Test_Recall@[20]_20.csv'])
```

```python
loss_normal_data = pd.read_csv(loss_normal)
loss_sorted_data = pd.read_csv(loss_sorted)
```

```python
x = loss_normal_data["Step"]
y1 = loss_normal_data["Value"]
y2 = loss_sorted_data["Value"]
fig, ax = plt.subplots(dpi=300)
ax.plot(x, y1, label='ML-1M normal')
ax.plot(x, y2, label='ML-1M sorted')
ax.set_yscale('log')
ax.set(xlabel='BPR', ylabel='BPRLoss')
ax.legend(loc='upper right')
ax.grid()
fig.savefig('../plots/loss_plot.eps', format='eps')
```
